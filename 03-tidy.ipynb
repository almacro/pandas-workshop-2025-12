```python
ebola = pd.read_csv("data/country_timeseries.csv")
```

```python
ebola
```

```python
ebola_long = ebola.melt(
    id_vars=["Date", "Day"],
    var_name="cd_country",
    value_name="count"
)
```

```python
ebola_long
```

```python
"Cases_Guinea".split("_")
```

```python
ebola_long.cd_country.str.split('_')
```

```python
ebola_long.cd_country.str.split('_', expand=True)
```

```python
ebola_long[["case_death", "country"]] = ebola_long.cd_country.str.split('_', expand=True)
```

```python
ebola_long
```

```python
ebola_long.groupby("case_death")["count"].max()
```

```python
ebola_long.groupby(["case_death", "country"])["count"].max()
```

```python
ebola_long.groupby(["case_death", "country"])["count"].max().reset_index()
```

```python
ebola_tidy = (ebola_long
 .drop(columns='cd_country')
 .pivot_table(
     index=['Date', 'Day', 'country'],
     columns='case_death',
     values='count')
             .reset_index())
ebola_tidy
```

```python
weather = pd.read_csv("data/weather.csv")
```

```python
weather
```

```python
weather_long = weather.melt(
    id_vars=["id", "year", "month", "element"],
    var_name="day",
    value_name="temp"
)
```

```python
weather_long
```

```python
weather_tidy = weather_long.pivot_table(
    index=["id", "year", "month", "day"],
    columns="element",
    values="temp"
)
weather_tidy
```

```python
weather_tidy.reset_index()
```

```python
weather_tidy.index
```

```python
weather_tidy = weather.melt(
    id_vars=["id", "year", "month", "element"],
    var_name="day",
    value_name="temp"
).pivot_table(
    index=["id", "year", "month", "day"],
    columns="element",
    values="temp"
).reset_index()

weather_tidy
```
