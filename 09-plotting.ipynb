```python
import plotnine as p9
from plotnine import ggplot, aes, geom_point
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
ggplot(tips, aes(x='total_bill', y='tip')) + geom_point()
```

```python
ggplot(tips, aes(x='total_bill', y='tip', color='sex')) + geom_point()
```

```python
ggplot(tips, aes(x='total_bill', y='tip', color='sex')) + geom_point() + p9.facet_wrap("size")
```

```python
ggplot(tips, aes(x='total_bill', y='tip', color='sex')) + geom_point() + p9.facet_grid("size", "day")
```

```python
(ggplot(tips, aes(x='total_bill', y='tip', color='sex'))
 + geom_point()
 + p9.facet_grid("size", "day")
 + p9.theme_xkcd()
)
```
